```python
# Sub-sample 100K sentences from a file
def sample_sentences(input_file_path, output_file_path):
    with open(input_file_path, 'r', encoding='utf-8') as file_in:
        sentences = file_in.readlines()[:100000]
    
    with open(output_file_path, 'w') as file_out:
        file_out.writelines(sentences)

# Sub-sample the English training data to 100K sentences
input_file_path = "/Users/songyafeng/6/Machine_Translation/Ex5/mt-exercise-5/data/train.en"
output_file_path = "/Users/songyafeng/6/Machine_Translation/Ex5/mt-exercise-5/train.en"

sample_sentences(input_file_path, output_file_path)
```

```python
# Sub-sample the Italian training data to 100K sentences
input_file_path = "/Users/songyafeng/6/Machine_Translation/Ex5/mt-exercise-5/data/train.it"
output_file_path = "/Users/songyafeng/6/Machine_Translation/Ex5/mt-exercise-5/train.it"

sample_sentences(input_file_path, output_file_path)
```
